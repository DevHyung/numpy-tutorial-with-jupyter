```python
# 아래의 결과가 나오도록 코드를 작성하세요 
# Hint : numpy의 타입? + plt.imshow() 사용
print(type(img))
plt.imshow(img)
```

```python
# 아래의 결과가 나오도록 코드를 작성하세요 
# Hint : 위의 사진색이 다르죠 ? 키워드는 BGR2RGB 
color_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# plt.imshow(img[:,:,::-1]) # Same = BGR -> RGB
plt.imshow(color_image)
```

```python
"""
사진이 가지고있는 RGB의 unique 값들을 출력
Hint : numpy + unique 
"""
np.unique(img)
```

```python
"""
copy_img 이미지를 
intensity 100 미만은 255으로 마스킹 후 출력
= 흑백반전
Hint : numpy masking
"""
offset = 100
copy_img[img <  offset] = 255
copy_img[img >= offset] = 0 
plt.imshow(copy_img)
```

```python
"""
copy_img 이미지를 
세로중 절반을 잘라 아래만출력해주세요
Hint : numpy slice
"""
h, w, c = copy_img.shape
plt.imshow(copy_img[int(h/2):, :])
```

```python
"""
copy_img 이미지를 
가로를 절반을 잘라 모두 출력해주세요
* 추가로 출력하면서 제목달아주면 굳
* 추가로 왼쪽이미지엔 H W 안보이게 해주면 굳 
Hint : plt subplot, numpy 
"""
h, w, c = copy_img.shape
left, right = np.split(copy_img, [int(w/2)], axis=1)

plt.subplot(1,2,1)
plt.imshow(left)
plt.title('LEFT')
plt.xticks([])
plt.yticks([])

plt.subplot(1,2,2)
plt.title('RIGHT')
plt.imshow(right)
```
